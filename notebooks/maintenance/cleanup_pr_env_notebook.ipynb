```python
from pyspark.sql import functions as F

# ---------- 1. Get pr_id parameter ----------
dbutils.widgets.text("pr_id", "")
pr_id = dbutils.widgets.get("pr_id")

assert pr_id, "pr_id is required (e.g. pr_123)"

print(f"Cleaning up environment for pr_id = '{pr_id}'")

# ---------- 2. Build DB names ----------
if pr_id == "prod":
    raise ValueError("Refusing to clean 'prod' environment via this notebook.")

raw_db_name = f"{pr_id}_raw"
clean_db_name = f"{pr_id}_clean"

print(f"Will try to drop schemas:")
print(f"  - {raw_db_name}")
print(f"  - {clean_db_name}")

# ---------- 3. Drop schemas if they exist ----------
for db_name in [raw_db_name, clean_db_name]:
    print(f"Dropping schema if exists: {db_name}")
    spark.sql(f"DROP SCHEMA IF EXISTS {db_name} CASCADE")

print("Cleanup completed âœ…")
```
